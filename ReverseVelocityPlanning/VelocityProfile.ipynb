```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
vel_prof = []
acceleration = 0.02
deceleration = 0.02
no_points = 100
top_speed = 2
```

```python
acc = [(acceleration * i) for i in range(no_points + 1)]
dec = [(top_speed - (deceleration * i)) for i in range(no_points + 1)]
const = [top_speed for i in range(2 * no_points + 1)]

vel_prof = acc + const + dec
```

```python
len(vel_prof)
```

```python
plt.plot(vel_prof)
```
