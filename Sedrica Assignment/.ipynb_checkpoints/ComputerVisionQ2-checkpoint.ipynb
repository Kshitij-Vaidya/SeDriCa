```python
from tensorflow.keras.datasets import cifar10

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
(X_Train, y_Train), (X_Test, y_Test) = cifar10.load_data()

print(X_Train.shape)
```

```python
# Normalising the training and testing set pixel values
X_Train, X_Test = X_Train / 255.0, X_Test / 255.0
```

# Corrected Model Implementation

The CNN architecture provided to us had a couple of issues that needed correction. Firstly the number of outputs was set to 12 but the 'cifar10' dataset only has 10 classes thus the output should be set to 10. Also the number of outputs of one layer need to match the number of inputs for the next layer. Another consideration is the choice of activation function for the Output layer. The given model used the Sigmoid Function which although possible is not the most optimal choice of function. The Softmax function generalises the Signoid for classifying multiple classes and does a better job here

We use the <b>tensorfklow.keras</b> Python library which has the functions required to create a Convolutional Network Model that will be used for Image Classification. The model used has 2 Convolutional Layers, 2 Max Pooling Layers and 2 Dense Node Layers. The Optimizer used is **Adam**, quite standard for these functionalities, the loss function evaluatued is Sparse Categorical Crossentropy. This function is used in multiple classification problems usually when the classes are mutually exclusive. The formula for **Sparse Categorical Cross-Entropy** takes into account the number of classes and if the sample belongs to the class or not.


```python
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential()

# Adding the Convolutional layers
model.add(layers.Conv2D(6, (5,5), 
                        activation='relu', padding='same',
                       input_shape=(X_Train.shape[1:])))
model.add(layers.MaxPooling2D(pool_size=(2, 2), strides=2))
model.add(layers.Conv2D(16, (5, 5), activation='relu'))
model.add(layers.MaxPooling2D((2,2), strides=2))

# Adding the Flatten layer
model.add(layers.Flatten())

# Adding Fully Connected Layers
model.add(layers.Dense(units=120, activation='relu', use_bias=True))
model.add(layers.Dense(units=84, activation='relu', use_bias=True))

# Output Layer
# There was a correction here, the output layer must have 10 outputs instead of 12 
# This is because of the number of classes of the Cifar10 dataset
# The activation function of the layer should also be Softmax and not Sigmoid
model.add(layers.Dense(10, activation='softmax'))

# Compiling the Model
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Training the Model over the Training Data
num_epochs = 20
batch_size = 64

Training = model.fit(X_Train, y_Train, 
                    batch_size=batch_size,
                    validation_data=(X_Test, y_Test),
                    epochs=num_epochs)
```

```python
plt.plot(Training.history['loss'], label='Training Loss',
        color='blue')
plt.plot(Training.history['val_loss'], label='Testing Loss',
         color='red')
plt.legend()
plt.xlabel('Number of Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
plt.plot(Training.history['accuracy'], label='Training Accuracy',
        color='blue')
plt.plot(Training.history['val_accuracy'], label='Testing Accuracy',
        color='red')
plt.legend()
plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy')
plt.show()
```

```python
Loss, Accuracy = model.evaluate(X_Test, y_Test)
print(f"Loss : {Loss}")
print(f"Accuracy : {Accuracy}")
```

```python
from sklearn.metrics import accuracy_score
y_Prob = model.predict(X_Test)
y_Pred = np.argmax(y_Prob, axis=1)

y_new_Test = y_Test

y_Test.flatten()

for i in range(10):
    IndexClass = np.where(y_new_Test == i)
    ClassAccuracy = accuracy_score(y_Test[IndexClass], y_Pred[IndexClass])
    print(f"Accuracy - Class {i} : {ClassAccuracy * 100.0}%")
```
